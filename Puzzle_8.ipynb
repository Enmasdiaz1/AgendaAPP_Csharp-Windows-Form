```python

#Librerías 

import argparse
import timeit
import resource
from collections import deque
from heapq import heappush, heappop, heapify
import itertools
```

```python
# clase estado

class State:

    def __init__(self, state, parent, move, depth, cost, key):

        self.state = state

        self.parent = parent

        self.move = move

        self.depth = depth

        self.cost = cost

        self.key = key

        if self.state:
            self.map = ''.join(str(e) for e in self.state)

    def __eq__(self, other):
        return self.map == other.map

    def __lt__(self, other):
        return self.map < other.map
```

```python
#Variables

goal_state = [0, 1, 2, 3, 4, 5, 6, 7, 8]
goal_node = State
initial_state = list()
board_len = 0
board_side = 0

nodes_expanded = 0
max_search_depth = 0
max_frontier_size = 0

moves = list()
costs = set()
```

```python
#Algoritmo bfs

"""Enmanuel Santos"""
def bfs(start_state):

#Declaramos variables glogables
    global max_frontier_size, goal_node, max_search_depth


#Seteando las variables, para que contengan los valores que se le asignen

    explored, queue = set(), deque([State(start_state, None, None, 0, 0, 0)])


#como es un bucle, hara esto hasta que sea diferente
    while queue:


      #sacando el primero que entro, para llegar al objetivo
        node = queue.popleft()

      #Añadiendo el nodo que ha sido explorado al mapa (Nodo Visitado)
        explored.add(node.map)

      #verificando si estan en el orden correcto, de ser asi--> "si el juego termino"
      #comparamos el Node.State (donde estamos) con goal_state (La Meta)

        if node.state == goal_state:
          #Asignamos el node que tenemos actualmente al goal_node y damos salida 
            goal_node = node
            #retornando al inicio deñ bucle
            return queue

        #exapndimos a los nuevos nodos y le asignamos ese valor a la variable "vecino" 
        #que como su nombre lo dice, se expande a lso nodos mas cercanos, o sea sus vecinos.
        
        neighbors = expand(node)

"""Francis T."""

        for neighbor in neighbors:
            #Si uno de los vecinos no se encuentra en el objeto explored pues lo agregamos al queue(se inserta por la parte derecha) y tambièn lo agregamos al objeto explored
            if neighbor.map not in explored:
                queue.append(neighbor)
                explored.add(neighbor.map)
                # Aqui realizamos una decision, si el neighbor.depth(que vendria siendo el camino que hemos trascurrido hasta llegar a nuestro objetivo) es mas alto que el max_search_depth(que vendria siendo la cantidad mas alta que se ha conseguido tras llegar a un objetivo) entonces se va a sumar 1 
                if neighbor.depth > max_search_depth:
                    max_search_depth += 1
        # Si tenemos mas elementos en el queue que en el max_frontier_size entonces habria un problema ya que se desbordaria nuestro tablero, es decir iria a lugares que no puedes explorar, entonces, para evitarlo decimos si la cantidad de elementos que hay en el queue es mas grande que la cantidad de elementos que hay en el max_frontier_size entonces se lo asignariamos a este ultimo
        if len(queue) > max_frontier_size:
            max_frontier_size = len(queue)
            
```

```python
#Algoritmo dfs

"""
Jeremy Arturo Herrera De Leon
Jean Carlos Reyes
"""

#El algoritmo dfs es un algoritmo que se encarga de buscar a profundidad todo para recorrer todos los nodos de manera ordenada

#La funcion dfs recibe el estado origen del arreglo
def dfs(start_state):

#En esta parte se declaran las variables globales, la maxima longitud de frontera, el noto meta y la profundidad de busqueda
    global max_frontier_size, goal_node, max_search_depth

#Aqui declaramos 2 variables, explored que recibira un dato vacio gracias a la funcion de python set y nos servira en el algoritmo para guardar los visitados,
#el stack que obtiene la lista de la clase State la cual tiene como parametros el estado inicial dado al ejecutar el programa
#el padre que en este caso es none,el movimiento none, la profundida 0, costo 0 y la llave 0.

    explored, stack = set(), list([State(start_state, None, None, 0, 0, 0)])

#Mientras la pila tenga dato entonces
    while stack:

#Agregaremos todo los datos y quitamos al que desapilamos con la funcion pop a la variable node que fue declarada aqui.
        node = stack.pop()
# Y se agrega el nodo explorado a la variable explored.
        explored.add(node.map)

#mientras mas profundo el nodo sera mas cerca del resultado
        if node.state == goal_state:
            goal_node = node
            return stack
#Aqui declaramos la variable neighbors que nos retornara el nodo invertido y expandido a la vez gracias a la funcion expand.
        neighbors = reversed(expand(node))

#Entonces iteramos con un for la lista de vecinos que nos retorno el nodo expandido..
        for neighbor in neighbors:
            #Si el vecino iterado no se encuentra en explorado pues los ponemos en la pila y lo agregamos a explorado.
            if neighbor.map not in explored:
                stack.append(neighbor)
                explored.add(neighbor.map)

#Se determina si la profundidad del vecino es mayor que la profundidad maxima de busqueda, de ser asi se incrementara en 1
                if neighbor.depth > max_search_depth:
                    max_search_depth += 1
#Si la longitud de la pila es mayor al tamaño maximo de la frontera entonces el tamaño maximo de la frontera tomara la longitud de la pila.
        if len(stack) > max_frontier_size:
            max_frontier_size = len(stack)
```

```python
#Algoritmo ast

"""
Luis Alberto Marmol
Yaher Hichez Garcia
"""

#El algoritmo A* o AST/A estrella es un algoritmo que trata de obtener el resultado a menor costo de distancia entre el nodo objetivo y el origen

#La funcion ast decibe el estado origen del arreglo
def ast(start_state):

#en esta parte se declaran las variables globales, la maxima longitud de frontera, el noto meta y la profundidad de busqueda
    global max_frontier_size, goal_node, max_search_depth

#se crea una lista para los campos explorados, dos para heapsort y un contador
    explored, heap, heap_entry, counter = set(), list(), {}, itertools.count()

#llave sera igual al estado origen
    key = h(start_state)

#se declara root y se le asigna una coleccion desde su estado inicial hasta los estados actuales
    root = State(start_state, None, None, 0, 0, key)

#Esta es la coleccion de entradas de los estados, los cuales son pasados para llegar al resultado
    entry = (key, 0, root)

    heappush(heap, entry)

#se procede a un mapeado a los estados segun su entrada
    heap_entry[root.map] = entry

#Este es el bucle de heapsearch, este lo compara y lo ordena
    while heap:
#aqui el nodo avanzara segun la coleccion de movimientos realizados
        node = heappop(heap)

#Los explorados seran agregados y apuntando la visita o movimiento realizado
        explored.add(node[2].map)

#mientras mas profundo el nodo sera mas cerca del resultado
        if node[2].state == goal_state:
            goal_node = node[2]
            return heap

#Los vecinos realizaran la comprobacion segun la profundidad o acercamiento y a su vez se determinara si no ha sido explorado
#Expand pertenece a sympy
        neighbors = expand(node[2])

        for neighbor in neighbors:

            neighbor.key = neighbor.cost + h(neighbor.state)

            entry = (neighbor.key, neighbor.move, neighbor)

            #Determinar si el vecino no fue explorado
            #Si no fue explorado 
            if neighbor.map not in explored:

                heappush(heap, entry)
#Si no ha sido explorado se agregara para posteriormente ser explorado

                explored.add(neighbor.map)

#Se mapeara los vecios en la coleccion de entrada o los movimientos hechos y asi validar la visita
                heap_entry[neighbor.map] = entry

#Se determina si la profundidad del vecino es mayor que la profundidad maxima de busqueda, de ser asi se incrementara en 1
                if neighbor.depth > max_search_depth:
                    max_search_depth += 1

#De lo contrario se mapeara el vecino para validad la entrada y saber si se ha encontrado una posicion correcta
            elif neighbor.map in heap_entry and neighbor.key < heap_entry[
                    neighbor.map][2].key:

                hindex = heap.index((heap_entry[neighbor.map][2].key,
                                     heap_entry[neighbor.map][2].move,
                                     heap_entry[neighbor.map][2]))

                heap[int(hindex)] = entry
                

                heap_entry[neighbor.map] = entry

                heapify(heap)
#Ya luego se evalua la coleccion para establecer la posicion correcta, y asi establecer la frontera y resolucion de busqueda
        if len(heap) > max_frontier_size:
            max_frontier_size = len(heap)
```

```python
#Expandir nodos

def expand(node):

    global nodes_expanded
    nodes_expanded += 1

    neighbors = list()

    neighbors.append(
        State(move(node.state, 1), node, 1, node.depth + 1, node.cost + 1, 0))
    neighbors.append(
        State(move(node.state, 2), node, 2, node.depth + 1, node.cost + 1, 0))
    neighbors.append(
        State(move(node.state, 3), node, 3, node.depth + 1, node.cost + 1, 0))
    neighbors.append(
        State(move(node.state, 4), node, 4, node.depth + 1, node.cost + 1, 0))

    nodes = [neighbor for neighbor in neighbors if neighbor.state]

    return nodes
```

```python
# movimientos

def move(state, position):

    #Copia el estado que recibbe
    new_state = state[:]

    index = new_state.index(0)
    #La posicion que recibe es un numero el cual, en funcion del numero que sea se mueve en una direccion.
    if position == 1:  # Up

        if index not in range(0, board_side):

            temp = new_state[index - board_side]
            new_state[index - board_side] = new_state[index]
            new_state[index] = temp

            return new_state
        else:
            return None

    if position == 2:  # Down

        if index not in range(board_len - board_side, board_len):

            temp = new_state[index + board_side]
            new_state[index + board_side] = new_state[index]
            new_state[index] = temp

            return new_state
        else:
            return None

    if position == 3:  # Left

        if index not in range(0, board_len, board_side):

            temp = new_state[index - 1]
            new_state[index - 1] = new_state[index]
            new_state[index] = temp

            return new_state
        else:
            return None

    if position == 4:  # Right

        if index not in range(board_side - 1, board_len, board_side):

            temp = new_state[index + 1]
            new_state[index + 1] = new_state[index]
            new_state[index] = temp

            return new_state
        else:
            return None
```

```python
# Posicion Manhattan

def h(state):

    #Calcula la suma total de todos los componentes o numeros que interfieren con una celda

    return sum(
        abs(b % board_side - g % board_side) +
        abs(b // board_side - g // board_side)
        for b, g in ((state.index(i), goal_state.index(i))
                     for i in range(1, board_len)))
```

```python
# backtrace

#La funcion backtrace lo que hace es que recorre el camino hacia atras cambiando los valores para que se puedan apreciar mejor a la hora de exportar 
def backtrace():

    current_node = goal_node

    while initial_state != current_node.state:

        #cambiando los valores de numericos a palabras
        if current_node.move == 1:
            movement = 'Up'
        elif current_node.move == 2:
            movement = 'Down'
        elif current_node.move == 3:
            movement = 'Left'
        else:
            movement = 'Right'

        moves.insert(0, movement)
        #Yendo de nodos hijos a nodos padres
        current_node = current_node.parent

    return moves
```

```python
# exportar

def export(frontier, time):

    global moves

    moves = backtrace()

    file = open('output.txt', 'w')
    file.write("path_to_goal: " + str(moves))
    file.write("\ncost_of_path: " + str(len(moves)))
    file.write("\nnodes_expanded: " + str(nodes_expanded))
    file.write("\nsearch_depth: " + str(goal_node.depth))
    file.write("\nmax_search_depth: " + str(max_search_depth))
    file.write("\nrunning_time: " + format(time, '.8f'))
    file.write("\nmax_ram_usage: " + format(
        resource.getrusage(resource.RUSAGE_SELF).ru_maxrss / 1000.0, '.8f'))
    file.close()
```

```python
# Leer

def read(configuration):

    global board_len, board_side

    data = configuration.split(",")

    for element in data:
        initial_state.append(int(element))

    board_len = len(initial_state)

    board_side = int(board_len**0.5)
```

```python
# clase principal

def main():

    parser = argparse.ArgumentParser()

    parser.add_argument('algorithm')
    parser.add_argument('board')
    args = parser.parse_args()

    read(args.board)

    function = function_map[args.algorithm]

    start = timeit.default_timer()

    frontier = function(initial_state)

    stop = timeit.default_timer()

    export(frontier, stop - start)
```
